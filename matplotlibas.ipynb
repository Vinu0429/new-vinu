```python
ml_student_age = np.random.randint(18,45, (100))
py_student_age =np.random.randint(15,40, (100))
```

```python
plt.hist(ml_student_age)
plt.xlabel("student category")
plt.ylabel("no student age")
plt.title("machine leran student")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype ="barstacked",align ="mid")
plt.xlabel("student category")
plt.ylabel("no student age")
plt.title("machine leran student")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype ="bar",align ="mid",orientation="horizontal",color = "m",label ="ml student")
plt.xlabel("student category")
plt.ylabel("no student age")
plt.title("machine leran student")
plt.legend(loc =4)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype ="bar",align ="mid",orientation="vertical",color = "m",label ="ml student")
plt.hist(py_student_age,bins,rwidth=0.8,histtype ="bar",align ="mid",orientation="vertical",color = "y",label ="py student")
plt.xlabel("student category")
plt.ylabel("no student age")
plt.title("machine leran student")
plt.legend(loc =3)
plt.show()
```

```python
classes = ["python ","ai","ml","ds","java"]
class_1 =[10,20,30,40,50]
class_2 =[20,38,10,30,55]
class_3 =[34,44,10,30,56]
```

```python
classes = ["python","machine leraning","artificial intelligence","data science","java"]
class_1 =[39,49,55,40,35]
explode = [0,0,0.4,0,0.2]  ## expolad means slice we can slice the part of diff float value 
colors = ["c","r","y","g","m"]
textprops ={"fontsize":15}
wedgeprops = {"linewidth":4, "width":3,"edgecolor":"k"}
plt.pie(
        class_1 , 
        labels =classes,
        explode = explode,
        colors=colors, 
        autopct = "%0.1f%%",
        shadow = True,
        radius = 1.4 ,
        startangle = 270,
        counterclock = True ,
        labeldistance=1.6 ,
        wedgeprops = wedgeprops,
        textprops =textprops,
        center =(2, 3),
        frame =True,
        rotatelabels = True 
        ) 
plt.show
```
